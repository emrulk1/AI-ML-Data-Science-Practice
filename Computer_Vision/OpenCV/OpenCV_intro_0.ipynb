# Import necessary modules

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

# Read Image using OpenCV

#### we read image using imread() method which takes 2 arguments : 
- image name
- read flag mode

#### Following read flag mode are used : 

- cv2.IMREAD_UNCHAGED => load image as it is

- cv2.IMREAD_GRAYSCALE => load in gray scale mode

- cv2.IMREAD_COLOR => load as colored image



```python
img = cv2.imread('./img/messi.jpeg', cv2.IMREAD_GRAYSCALE)
print(img)
```
