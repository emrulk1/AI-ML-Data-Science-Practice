```python
def show_random_part() :
    import random
    xlim , ylim , _ = img_np.shape
    x1 = random.randint(0, xlim)
    x2 = random.randint(0, xlim)
    y1 = random.randint(0, ylim)
    y2 = random.randint(0, ylim)

    
    if x1 > x2 :
        #swap x1,x2
        x1 , x2 = x2 , x1

    print("x1 = ", x1)
    print("x2 = ", x2)


    if y1 > y2 : 
        #swap y1,y2
        y1 , y2 = y2 , y1

    print("y1 = ", y1)
    print("y2 = ", y2)

    plt.imshow(img_np[x1:x2, y1:y2])
    plt.show()
```

```python
for i in range(5) :
    show_random_part()
```
