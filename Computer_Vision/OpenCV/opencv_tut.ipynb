```python
import cv2 as cv
import numpy as np
 
# Load two images
img1 = cv.imread('img/satellite-image-of-globe.jpg')
img2 = cv.imread('img/eso1907a.jpg')

# take a crop of first image with size of second image as ROI to work on
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols] #ROI stands for region of interest

# create a mask of logo and create its inverse mask also
img2gray = cv.cvtColor(img2,cv.COLOR_BGR2GRAY) #convert image to grayscale
ret, mask = cv.threshold(img2gray, 10, 255, cv.THRESH_BINARY) #create a mask by thresholding grayscale image
mask_inv = cv.bitwise_not(mask) #inverse the mask

# Now black-out the area of mask in ROI
img1_bg = cv.bitwise_and(roi,roi,mask = mask_inv) #and operation on roi using mask to remove only mask area

# Take only region of mask from logo image.
img2_fg = cv.bitwise_and(img2,img2,mask = mask) #and operation on second image using inverse of mask to keep only mask area

# Put masked area of second image in ROI and modify the main image
dst = cv.add(img1_bg,img2_fg) #add ROI with mask area removed and second image with only masked area (looks like only masked area is replaced)
img1[0:rows, 0:cols ] = dst #replcae ROI in first image (result is first image is partially replaced with second image)

cv.imshow('res',img1) #show image using opencv
cv.waitKey(0) #wait indefinately 
cv.destroyAllWindows() #destroy all windows opened by opencv
```

```python
#contour approximation (approximates contours to another shape with minimum vartices)
img = cv2.imread('img/k_sq.png')
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thr = cv2.threshold(imgray,127,255,0)
contr, heirarchy = cv2.findContours(thr, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
imgd = cv2.drawContours(img.copy(),contr,-1,(0,0,250),5)
epsilon = .1*cv2.arcLength(contr[0],True) #calculate epsilon for approximation from contour perimeter
contr2 = cv2.approxPolyDP(contr[0],epsilon,True) #approximate contour usign epsilon calculated from perimeter
imgd2 = cv2.drawContours(img.copy(),[contr2],0,(0,0,250),5) #draw the approximated contour

plt.subplot(121), plt.imshow(imgd)
plt.subplot(122), plt.imshow(imgd2)
plt.show()
```


connectedComponentsWithStats():

function is used to find all connected components and statistics of those components. statistics include cv2.CC_STAT_LEFT - The leftmost (x) coordinate which is the inclusive start of the bounding box in the horizontal direction. cv2.CC_STAT_TOP - The topmost (y) coordinate which is the inclusive start of the bounding box in the vertical direction. cv2.CC_STAT_WIDTH - The horizontal size of the bounding box. cv2.CC_STAT_HEIGHT - The vertical size of the bounding box. cv2.CC_STAT_AREA - The total area (in pixels) of the connected component.
Function Syntax

retval, labels, stats, centroids = cv.connectedComponentsWithStats(image[, labels[, stats[, centroids[, connectivity[, ltype]]]]])

Parameters:
    image - the 8-bit single-channel image to be labeled
    labels - destination labeled image (optional)
    stats - optional (description at returned stats)
    centroids - optional (description same as returun centroids)
    connectivity - 8 or 4 for 8-way or 4-way connectivity respectively (optional)
    ltype - output image label type. Currently CV_32S and CV_16U are supported. (optional)

Returns:
    retval - N, number of connected components(labels) found. [0 - N-1], 0 is background

    labels - returned image having all the connected components numbered [0 - N-1].
              each pixel under a connected component(n) is valued n (number of that connected component)

    stats - statistics output for each label, including the background label. 
              Statistics are accessed via stats(label, COLUMN) where, 
              COLUMN is one of ConnectedComponentsTypes, selecting the 
              statistic. The data type is CV_32S. ConnectedComponentsTypes are 
              cv2.CC_STAT_LEFT, cv2.CC_STAT_TOP, cv2.CC_STAT_WIDTH, cv2.CC_STAT_HEIGHT, cv2.CC_STAT_AREA

    centroids - centroid output for each label, including the background label. 
                  Centroids are accessed via centroids(label, 0) for x and centroids(label, 1) for y. 
                  The data type CV_64F.



```python
#remove connected components with small area
img = cv2.imread('img/img.png')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,img2 = cv2.threshold(imgray,127,255,cv2.THRESH_BINARY)

#get connected components and their statistics
nlabels, labels, stats, centroids = cv2.connectedComponentsWithStats(img2)

#select area of every connected component except for background
area = stats[1:, cv2.CC_STAT_AREA] #flags can be cv2.CC_STAT_*, *=TOP/LEFT/HEIGHT/WIDTH/AREA
nlabels = nlabels - 1
img3 = img2.copy()

min_size = 600
# for every component in the image keep it only if it's above min_size
for i in range(0, nlabels):
    if area[i] < min_size:
        img3[labels == i+1] = 0
plt.subplot(221), plt.imshow(img), plt.title('original')
plt.subplot(222), plt.imshow(img2,'gray'), plt.title('thresholded')
plt.subplot(223), plt.imshow(labels, 'gray'), plt.title('labels (connectedComponents)')
plt.subplot(224), plt.imshow(img3, 'gray'), plt.title('small components removed')
plt.show()
```
